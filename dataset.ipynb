## Data Cleaning and Preprocessing for Lending Club Dataset

Now let's perform comprehensive data cleaning and feature engineering for credit risk modeling.

```python
# Data Cleaning and Preprocessing for Lending Club Dataset
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

print("=== DATA CLEANING AND PREPROCESSING ===\n")

# Create a working copy of the dataset for cleaning
# Given the large size, let's sample the data for efficient processing
print("Due to dataset size (2.2M+ records), sampling 200,000 records for analysis...")

# Set random seed for reproducibility
np.random.seed(42)

# Stratified sampling to maintain default rate distribution
n_samples = 200000
good_loans = df_lend[df_lend['is_default'] == 0]
bad_loans = df_lend[df_lend['is_default'] == 1]

# Calculate proportional sample sizes
default_rate = df_lend['is_default'].mean()
n_bad = int(n_samples * default_rate)
n_good = n_samples - n_bad

# Sample from each group
good_sample = good_loans.sample(n=n_good, random_state=42)
bad_sample = bad_loans.sample(n=n_bad, random_state=42)

# Combine samples
df_sample = pd.concat([good_sample, bad_sample]).sample(frac=1, random_state=42).reset_index(drop=True)

print(f"Sample dataset created:")
print(f"  Total records: {len(df_sample):,}")
print(f"  Good loans: {(df_sample['is_default'] == 0).sum():,} ({(df_sample['is_default'] == 0).mean():.1%})")
print(f"  Bad loans: {(df_sample['is_default'] == 1).sum():,} ({(df_sample['is_default'] == 1).mean():.1%})")

print("\n=== DATA QUALITY ANALYSIS ===")

# Select features for modeling (removing obvious leakage features)
modeling_features = [
    # Loan characteristics
    'loan_amnt', 'term', 'int_rate', 'grade', 'sub_grade',
    
    # Borrower characteristics  
    'emp_length', 'home_ownership', 'annual_inc', 'verification_status',
    'purpose', 'dti', 
    
    # Credit history
    'delinq_2yrs', 'earliest_cr_line', 'inq_last_6mths', 'open_acc', 
    'pub_rec', 'revol_bal', 'revol_util', 'total_acc',
    
    # Application details
    'initial_list_status', 'application_type', 'mort_acc', 'pub_rec_bankruptcies'
]

# Filter to available features
available_modeling_features = [feat for feat in modeling_features if feat in df_sample.columns]
print(f"Features selected for modeling: {len(available_modeling_features)}")

# Create clean dataset with selected features + target
df_clean = df_sample[available_modeling_features + ['is_default']].copy()

print(f"\nCleaned dataset shape: {df_clean.shape}")

# Analyze missing values in key features
print(f"\nMissing values analysis:")
missing_analysis = df_clean.isnull().sum()
missing_pct = (missing_analysis / len(df_clean)) * 100

missing_summary = pd.DataFrame({
    'Missing_Count': missing_analysis,
    'Missing_Percentage': missing_pct
}).sort_values('Missing_Percentage', ascending=False)

print(missing_summary[missing_summary['Missing_Count'] > 0].round(2))
```

```python
# Feature Engineering and Data Cleaning
print("=== FEATURE ENGINEERING ===\n")

# Handle missing values and data quality issues
df_engineered = df_clean.copy()

# 1. Clean and transform key features

# Clean DTI ratio (remove outliers and invalid values)
print("1. CLEANING DTI RATIO...")
dti_outliers = df_engineered['dti'] > 100  # DTI > 100% is unrealistic
print(f"  DTI outliers (>100%): {dti_outliers.sum()} records")
df_engineered.loc[dti_outliers, 'dti'] = np.nan

# Fill missing DTI with median
dti_median = df_engineered['dti'].median()
df_engineered['dti'].fillna(dti_median, inplace=True)

# 2. Interest rate processing
print("2. PROCESSING INTEREST RATE...")
# Create interest rate bins
df_engineered['int_rate_bin'] = pd.cut(df_engineered['int_rate'], 
                                      bins=[0, 7, 10, 13, 16, 20, 35],
                                      labels=['Very Low', 'Low', 'Medium', 'High', 'Very High', 'Extreme'])

# High interest rate flag
df_engineered['high_int_rate'] = (df_engineered['int_rate'] > 15).astype(int)

# 3. Employment length processing
print("3. PROCESSING EMPLOYMENT LENGTH...")
# Convert employment length to numeric
emp_length_map = {
    'n/a': 0,
    '< 1 year': 0.5,
    '1 year': 1,
    '2 years': 2,
    '3 years': 3,
    '4 years': 4,
    '5 years': 5,
    '6 years': 6,
    '7 years': 7,
    '8 years': 8,
    '9 years': 9,
    '10+ years': 10
}

df_engineered['emp_length_numeric'] = df_engineered['emp_length'].map(emp_length_map)
df_engineered['emp_length_numeric'].fillna(0, inplace=True)  # Fill missing with 0

# Employment stability flags
df_engineered['stable_employment'] = (df_engineered['emp_length_numeric'] >= 3).astype(int)
df_engineered['new_employee'] = (df_engineered['emp_length_numeric'] < 1).astype(int)

# 4. Annual income processing
print("4. PROCESSING ANNUAL INCOME...")
# Handle extreme income outliers
income_99th = df_engineered['annual_inc'].quantile(0.99)
print(f"  99th percentile income: ${income_99th:,.0f}")

# Cap income at 99th percentile
df_engineered.loc[df_engineered['annual_inc'] > income_99th, 'annual_inc'] = income_99th

# Fill missing income with median
income_median = df_engineered['annual_inc'].median()
df_engineered['annual_inc'].fillna(income_median, inplace=True)

# Income-based features
df_engineered['income_to_loan_ratio'] = df_engineered['annual_inc'] / df_engineered['loan_amnt']
df_engineered['high_income'] = (df_engineered['annual_inc'] > 100000).astype(int)
df_engineered['low_income'] = (df_engineered['annual_inc'] < 40000).astype(int)

# 5. Credit utilization processing
print("5. PROCESSING REVOLVING UTILIZATION...")
# Clean revolving utilization
df_engineered['revol_util'] = df_engineered['revol_util'].fillna(0)

# Handle outliers in revolving utilization
df_engineered.loc[df_engineered['revol_util'] > 100, 'revol_util'] = 100

# Utilization risk features
df_engineered['high_utilization'] = (df_engineered['revol_util'] > 80).astype(int)
df_engineered['maxed_out'] = (df_engineered['revol_util'] >= 95).astype(int)

# 6. Credit history length
print("6. PROCESSING CREDIT HISTORY...")
# Convert earliest_cr_line to years of credit history
current_year = 2018  # Dataset goes to 2018Q4
df_engineered['earliest_cr_line'] = pd.to_datetime(df_engineered['earliest_cr_line'])
df_engineered['credit_history_years'] = current_year - df_engineered['earliest_cr_line'].dt.year

# Handle missing values
df_engineered['credit_history_years'].fillna(df_engineered['credit_history_years'].median(), inplace=True)

# Credit history features
df_engineered['thin_credit_file'] = (df_engineered['credit_history_years'] < 5).astype(int)
df_engineered['established_credit'] = (df_engineered['credit_history_years'] > 10).astype(int)

# 7. Derogatory records
print("7. PROCESSING DEROGATORY RECORDS...")
# Fill missing values
for col in ['delinq_2yrs', 'pub_rec', 'pub_rec_bankruptcies']:
    if col in df_engineered.columns:
        df_engineered[col].fillna(0, inplace=True)

# Derogatory flags
df_engineered['has_delinq'] = (df_engineered['delinq_2yrs'] > 0).astype(int)
df_engineered['has_pub_rec'] = (df_engineered['pub_rec'] > 0).astype(int)
if 'pub_rec_bankruptcies' in df_engineered.columns:
    df_engineered['has_bankruptcy'] = (df_engineered['pub_rec_bankruptcies'] > 0).astype(int)

# 8. Loan characteristics
print("8. CREATING LOAN CHARACTERISTIC FEATURES...")
# Term processing
df_engineered['term_numeric'] = df_engineered['term'].str.extract('(\d+)').astype(float)

# Large loan flag
loan_75th = df_engineered['loan_amnt'].quantile(0.75)
df_engineered['large_loan'] = (df_engineered['loan_amnt'] > loan_75th).astype(int)

# Grade-based features
high_risk_grades = ['E', 'F', 'G']
df_engineered['high_risk_grade'] = df_engineered['grade'].isin(high_risk_grades).astype(int)

# 9. Create composite risk scores
print("9. CREATING COMPOSITE RISK SCORES...")

# Credit risk score based on credit history
credit_risk_factors = ['has_delinq', 'has_pub_rec', 'high_utilization', 'thin_credit_file']
if 'has_bankruptcy' in df_engineered.columns:
    credit_risk_factors.append('has_bankruptcy')

df_engineered['credit_risk_score'] = df_engineered[credit_risk_factors].sum(axis=1)

# Financial stress score
financial_stress_factors = ['high_int_rate', 'high_risk_grade', 'low_income', 'new_employee']
df_engineered['financial_stress_score'] = df_engineered[financial_stress_factors].sum(axis=1)

# Overall risk score
df_engineered['total_risk_score'] = df_engineered['credit_risk_score'] + df_engineered['financial_stress_score']

print("\n=== FEATURE ENGINEERING SUMMARY ===")
original_features = len([col for col in df_engineered.columns if col in df_clean.columns])
new_features = len([col for col in df_engineered.columns if col not in df_clean.columns])

print(f"Original features: {original_features}")
print(f"New engineered features: {new_features}")
print(f"Total features: {len(df_engineered.columns)}")

# Show new feature names
new_feature_names = [col for col in df_engineered.columns if col not in df_clean.columns]
print(f"\nNew engineered features ({len(new_feature_names)}):")
for i, feature in enumerate(new_feature_names, 1):
    print(f"{i:2d}. {feature}")

print(f"\nFinal dataset shape: {df_engineered.shape}")

# Check for remaining missing values
remaining_missing = df_engineered.isnull().sum().sum()
print(f"Remaining missing values: {remaining_missing}")
```

```python
# Final Data Preparation for Modeling
print("=== FINAL DATA PREPARATION ===")

# Start with df_clean which has the target variable 'is_default'
df_final = df_clean.copy()

# Apply key feature engineering
print("Applying feature engineering...")

# 1. Clean DTI ratio
dti_outliers = df_final['dti'] > 100
print(f"DTI outliers (>100%): {dti_outliers.sum()} records")
df_final.loc[dti_outliers, 'dti'] = np.nan

# 2. Interest rate features
df_final['high_int_rate'] = (df_final['int_rate'] > 15).astype(int)

# 3. Employment length processing
emp_length_map = {
    'n/a': 0, '< 1 year': 0.5, '1 year': 1, '2 years': 2, '3 years': 3, 
    '4 years': 4, '5 years': 5, '6 years': 6, '7 years': 7, '8 years': 8, 
    '9 years': 9, '10+ years': 10
}
df_final['emp_length_numeric'] = df_final['emp_length'].map(emp_length_map)
df_final['stable_employment'] = (df_final['emp_length_numeric'] >= 3).astype(int)

# 4. Income features
income_99th = df_final['annual_inc'].quantile(0.99)
df_final.loc[df_final['annual_inc'] > income_99th, 'annual_inc'] = income_99th
df_final['income_to_loan_ratio'] = df_final['annual_inc'] / df_final['loan_amnt']
df_final['high_income'] = (df_final['annual_inc'] > 100000).astype(int)

# 5. Credit utilization
df_final['revol_util'] = df_final['revol_util'].fillna(0)
df_final.loc[df_final['revol_util'] > 100, 'revol_util'] = 100
df_final['high_utilization'] = (df_final['revol_util'] > 80).astype(int)

# 6. Create risk indicators
df_final['has_delinq'] = (df_final['delinq_2yrs'] > 0).astype(int)
df_final['has_pub_rec'] = (df_final['pub_rec'] > 0).astype(int)

# Large loan flag
loan_75th = df_final['loan_amnt'].quantile(0.75)
df_final['large_loan'] = (df_final['loan_amnt'] > loan_75th).astype(int)

# High-risk grade
high_risk_grades = ['E', 'F', 'G']
df_final['high_risk_grade'] = df_final['grade'].isin(high_risk_grades).astype(int)

print("Feature engineering completed!")

# Handle missing values
print("\nHandling missing values...")

# For numeric columns
numeric_cols = df_final.select_dtypes(include=[np.number]).columns
for col in numeric_cols:
    if df_final[col].isnull().sum() > 0:
        median_val = df_final[col].median()
        df_final[col].fillna(median_val, inplace=True)
        print(f"  Filled {col} missing values with median: {median_val}")

# For categorical columns
categorical_cols = df_final.select_dtypes(include=['object']).columns
for col in categorical_cols:
    if df_final[col].isnull().sum() > 0:
        mode_val = df_final[col].mode()
        if len(mode_val) > 0:
            df_final[col].fillna(mode_val.iloc[0], inplace=True)
            print(f"  Filled {col} missing values with mode: {mode_val.iloc[0]}")

# Check remaining missing values
final_missing = df_final.isnull().sum().sum()
print(f"\nRemaining missing values: {final_missing}")

# Prepare features for modeling
exclude_features = ['is_default', 'earliest_cr_line']  # 'is_default' is our target
feature_columns = [col for col in df_final.columns if col not in exclude_features]

X = df_final[feature_columns].copy()
y = df_final['is_default'].copy()

print(f"\n=== DATASET READY FOR MODELING ===")
print(f"Features: {X.shape[1]}")
print(f"Samples: {X.shape[0]:,}")
print(f"Target distribution:")
print(f"  No default (0): {(y == 0).sum():,} ({(y == 0).mean():.1%})")
print(f"  Default (1): {(y == 1).sum():,} ({(y == 1).mean():.1%})")

# Show feature types
print(f"\nFeature types:")
dtype_summary = X.dtypes.value_counts()
for dtype, count in dtype_summary.items():
    print(f"  {dtype}: {count} features")

print(f"\nSample feature list (first 10):")
for i, col in enumerate(feature_columns[:10], 1):
    print(f"  {i:2d}. {col} ({df_final[col].dtype})")

print("..." if len(feature_columns) > 10 else "")
print(f"Dataset preparation completed successfully!")
```

```python
# Machine Learning Pipeline for Lending Club Credit Risk
print("=== MACHINE LEARNING PIPELINE ===\n")

from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
import xgboost as xgb
from lightgbm import LGBMClassifier
import warnings
warnings.filterwarnings('ignore')

# Prepare data for modeling
print("1. PREPARING DATA FOR MODELING...")

# Encode categorical variables
X_encoded = X.copy()
label_encoders = {}

categorical_columns = X_encoded.select_dtypes(include=['object']).columns
print(f"Encoding {len(categorical_columns)} categorical features...")

for col in categorical_columns:
    le = LabelEncoder()
    X_encoded[col] = le.fit_transform(X_encoded[col].astype(str))
    label_encoders[col] = le
    print(f"  {col}: {len(le.classes_)} unique categories")

print(f"Final feature matrix shape: {X_encoded.shape}")
print(f"All features are now numeric: {X_encoded.dtypes.nunique() <= 2}")  # Should be only int64 and float64

# Split the data
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42, stratify=y
)

print(f"\nData split completed:")
print(f"  Training set: {X_train.shape[0]:,} samples")
print(f"  Test set: {X_test.shape[0]:,} samples")
print(f"  Training default rate: {y_train.mean():.1%}")
print(f"  Test default rate: {y_test.mean():.1%}")

# Scale features for logistic regression
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Feature scaling completed for logistic regression.")
```

```python
# Model Evaluation Function
def evaluate_model(model, X_train, X_test, y_train, y_test, model_name):
    """Comprehensive model evaluation function"""
    print(f"\n=== {model_name.upper()} MODEL EVALUATION ===")
    
    # Train the model
    print("Training model...")
    model.fit(X_train, y_train)
    
    # Make predictions
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    y_train_proba = model.predict_proba(X_train)[:, 1] if hasattr(model, 'predict_proba') else None
    y_test_proba = model.predict_proba(X_test)[:, 1] if hasattr(model, 'predict_proba') else None
    
    # Calculate metrics
    from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
    
    train_accuracy = accuracy_score(y_train, y_train_pred)
    test_accuracy = accuracy_score(y_test, y_test_pred)
    train_auc = roc_auc_score(y_train, y_train_proba) if y_train_proba is not None else None
    test_auc = roc_auc_score(y_test, y_test_proba) if y_test_proba is not None else None
    
    precision = precision_score(y_test, y_test_pred)
    recall = recall_score(y_test, y_test_pred)
    f1 = f1_score(y_test, y_test_pred)
    
    print(f"Training Accuracy: {train_accuracy:.4f}")
    print(f"Test Accuracy: {test_accuracy:.4f}")
    if train_auc and test_auc:
        print(f"Training AUC: {train_auc:.4f}")
        print(f"Test AUC: {test_auc:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-Score: {f1:.4f}")
    
    # Classification report
    print(f"\nClassification Report:")
    print(classification_report(y_test, y_test_pred))
    
    # Store results
    results = {
        'model_name': model_name,
        'model': model,
        'train_accuracy': train_accuracy,
        'test_accuracy': test_accuracy,
        'train_auc': train_auc,
        'test_auc': test_auc,
        'precision': precision,
        'recall': recall,
        'f1_score': f1,
        'y_test_pred': y_test_pred,
        'y_test_proba': y_test_proba
    }
    
    return results

# 2. TRAIN AND EVALUATE MODELS
print("\n2. TRAINING AND EVALUATING MODELS...")

model_results = []

# Logistic Regression
print("\n" + "="*50)
lr_model = LogisticRegression(random_state=42, max_iter=1000)
lr_results = evaluate_model(lr_model, X_train_scaled, X_test_scaled, y_train, y_test, "Logistic Regression")
model_results.append(lr_results)
```

```python
# Decision Tree
print("\n" + "="*50)
dt_model = DecisionTreeClassifier(random_state=42, max_depth=10, min_samples_split=100)
dt_results = evaluate_model(dt_model, X_train, X_test, y_train, y_test, "Decision Tree")
model_results.append(dt_results)

# Random Forest
print("\n" + "="*50)
rf_model = RandomForestClassifier(random_state=42, n_estimators=100, max_depth=10, min_samples_split=100, n_jobs=-1)
rf_results = evaluate_model(rf_model, X_train, X_test, y_train, y_test, "Random Forest")
model_results.append(rf_results)

# XGBoost
print("\n" + "="*50)
xgb_model = xgb.XGBClassifier(random_state=42, n_estimators=100, max_depth=6, learning_rate=0.1, 
                              eval_metric='logloss', verbosity=0)
xgb_results = evaluate_model(xgb_model, X_train, X_test, y_train, y_test, "XGBoost")
model_results.append(xgb_results)

# LightGBM
print("\n" + "="*50)
lgb_model = LGBMClassifier(random_state=42, n_estimators=100, max_depth=6, learning_rate=0.1, 
                           verbosity=-1, force_col_wise=True)
lgb_results = evaluate_model(lgb_model, X_train, X_test, y_train, y_test, "LightGBM")
model_results.append(lgb_results)

print("\n" + "="*50)
print("ALL MODELS TRAINED SUCCESSFULLY!")
print("="*50)
```
